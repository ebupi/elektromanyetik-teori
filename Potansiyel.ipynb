## Elektromagnetik Teori Potansiyel Çizim Ödevi

 $V(x,y) = \dfrac{4V_0}{\pi} \sum_{k = 1,3,5,\dots} \dfrac{1}{k} e^{-kx} \sin ky$ denklemine ait üç boyutlu yüzeyi elde etmek için yazılmıştır.

```python
# kütüphaneleri içe aktarma
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# potansiyel fonksiyonunu tanımlama
def potential(x, y, V0):
    # Fourier serisi için toplamı başlatma
    total = 0
    # k = 1, 3, 5, ... için döngüye alma
    k = 1
    while True:
        # her terimin değerini hesaplama
        term = (4*V0/np.pi) * (1/k) * np.exp(-k*x) * np.sin(k*y)
        # toplama ekleme
        total += term
        # toplamın mutlak değeri 1'den küçükse döngüyü durdur
        if np.all(np.abs(total) < 1):
            break
        # k değerini 2 artırma
        k += 2
    # Toplamı döndürme
    return total

# x ve y değerleri için aralıkları belirleme
x_min = 0
x_max = 3
y_min = 0
y_max = 3

# x ve y değerleri için adımları belirleme
x_step = 0.1
y_step = 0.1

# x ve y değerleri için dizileri oluşturma
x = np.arange(x_min, x_max + x_step, x_step)
y = np.arange(y_min, y_max + y_step, y_step)

# x ve y değerleri için ızgarayı oluşturma
X, Y = np.meshgrid(x, y)

# V0 değerini belirleme
V0 = 1

# z değerlerini hesapla
Z = potential(X, Y, V0)

# 3 boyutlu grafik oluştur
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

# yüzey grafiğini çiz
ax.plot_surface(Z, Y, X, cmap='gray')
ax2.plot_surface(Z, Y, X, cmap='gray')

# görüş açısını ayarla
ax.view_init(elev=30, azim=30)
# görüşü yandan görünecek şekilde ayarla
ax2.view_init(elev=30, azim=90)

# ilk grafik başlığı ve eksin isimleri
ax.set_title(f"$V(x,y)$ Şekil 3.17")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel(f"$V(x,y)$")

# ikinci grafik başlığı ve eksin isimleri
ax2.set_title(f"$V(x,y)$ Şekil 3.17 yandan görünüm")
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel(f"$V(x,y)$")

# grafik boyutu
fig.set_size_inches(8, 8)

# grafiği kaydet
# plt.savefig('potansiyel-cizim-2.png')

# grafiği gösterme
plt.show()
```
