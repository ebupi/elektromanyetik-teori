Türkçe kitap : $V(x,y) = \dfrac{4V_0}{\pi} \sum_{k = 1,3,5,\dots} \dfrac{1}{k} e^{-kx} \sin ky$

İngilizce kitap : $V(x,y) = \dfrac{4V_0}{\pi} \sum_{n = 1,3,5,\dots} e^{-n \pi x/a} \sin (n \pi y/a)$

Bu denklemden yola çıkarak grafiği çizmek için x,y,z değerlerine ihtiyacımız var;

+ V(x,y) = z (döngü içinden çıkan değerler)
+ x = x (0 ile 600) alsam deneme amaçlı
+ y = y (0 ile 2$\pi$) arasında alacağım

```python
# kendi yaklaşımım
def potansiyel_plot():
    """$V(x,y) = \dfrac{4V_0}{\pi} \sum_{k = 1,3,5,\dots} \dfrac{1}{k} e^{-kx} \sin ky$ 
    yukardaki denkleme ait potansiyel yüzeyi oluşturma ve çizdirmek için yazılmıştır."""
    
    v_0 = 1 # ilk potansiyel
    X = np.arange(0,150) # x değerlerini içeren numpy arrayi
    Y = np.linspace(0, 2*np.pi, num=150) # y değerlerini içeren numpy değerleri
    k = np.arange(1, 251, 2) # k tek sayılarını içren numpy arrayi
    print(len(X),len(Y)) # x ve y alanlarının uzunlukları birbirine eşit mi?
    V = np.zeros_like(X, dtype=float) # döngünün içinden çıkan V(x,y) potansiyel değerleri
    print(len(V)) # z ekseni V(x,y) değerleri uzunlukları birbirine eşit mi?
    
    for k in k:
        V += 4*(v_0/np.pi)*(1/k)*np.exp(-k*X)*np.sin(k*Y)   # döngü V değerlerini hespla ve arrayi güncelle

    
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111, projection='3d')
    
    # 3 boyutlu bir grafik oluşturma
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # yüzey grafiğini çizme
    ax.plot_surface(X, Y, V, cmap='tab20b')

    # görüş açısını ayarla
    #ax.view_init(elev=30, azim=30)

    # grafik başlığı ve eksen isimleri
    ax.set_title('FIGURE 3.17')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    # y eksenini ters çevir
    #plt.gca().invert_yaxis()
    #plt.gca().invert_xaxis()
    #plt.gca().invert_zaxis()
    ax.set_zlabel('V/V0')
    ax.set_zlim(0,1)


    # grafiği göster
    plt.show()
```

```python
potansiyel_plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# gpt + benim yaptıklarım

def potansiyel_grafik():
    v_0 = 1
    x = np.arange(0, 600, 5)  # 0 ile 600 arasında 5 birimlik adımlarla
    y = np.linspace(0, 2*np.pi, num=100)
    
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X, dtype=float)
    
    for k in range(1, 100, 2):  # Tek sayılar için
        Z += (4*v_0/np.pi) * (1/k) * np.exp(-k*X) * np.sin(k*Y)
    
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Çizim
    surface = ax.plot_surface(X, Y, Z, cmap='viridis', rstride=1, cstride=1, alpha=0.8, linewidth=0.5, edgecolors='k')

    # Eksen isimleri
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('V(x,y)')
    
    # Başlık
    ax.set_title('Potansiyel Grafiği')

    # Renk çubuğu
    fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

    # Göster
    plt.show()

# Fonksiyonu çağır
potansiyel_grafik()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def V(x, y, V0, k_max):
    result = 0
    for k in range(1, k_max + 1):
        result += (1/k) * np.exp(-k*x) * np.sin(k*y)
    return (4*V0/np.pi) * result

# Parametreler
V0 = 1.0
k_max = 100

# X, Y ve Z değerlerini oluştur
x_values = np.linspace(0, 5, 100)
y_values = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x_values, y_values)
Z = V(X, Y, V0, k_max)

# 3D grafiği çiz
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Eksen etiketlerini ekleyin
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('V(x, y)')

# Grafiği göster
plt.show()
```
